```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def plot_surface(X, Y, Z, title="3D Shape", cmap=cm.viridis):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap=cmap, edgecolor='none', alpha=0.9)
    ax.set_title(title)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.tight_layout()
    plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
u, v = np.meshgrid(u, v)

X = np.cos(u) * np.sin(v)
Y = np.sin(u) * np.sin(v)
Z = np.cos(v)

plot_surface(X, Y, Z, "üåç Sphere")
```

```python
R = 1  # major radius
r = 0.3  # minor radius

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

X = (R + r * np.cos(v)) * np.cos(u)
Y = (R + r * np.cos(v)) * np.sin(u)
Z = r * np.sin(v)

plot_surface(X, Y, Z, "üç© Torus")
```

```python
theta = np.linspace(0, 2 * np.pi, 150)
w = np.linspace(-0.3, 0.3, 30)
theta, w = np.meshgrid(theta, w)

R = 1
X = (R + w * np.cos(theta / 2)) * np.cos(theta)
Y = (R + w * np.cos(theta / 2)) * np.sin(theta)
Z = w * np.sin(theta / 2)

plot_surface(X, Y, Z, "‚ôæÔ∏è M√∂bius Strip")
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)

z = x**2 - y**2

plot_surface(x, y, z, "üß™ Hyperbolic Paraboloid (Saddle)")
```

```python
u = np.linspace(-1.5, 1.5, 100)
v = np.linspace(-1.5, 1.5, 100)
u, v = np.meshgrid(u, v)

X = u - (u**3)/3 + u*v**2
Y = v - (v**3)/3 + v*u**2
Z = u**2 - v**2

plot_surface(X, Y, Z, "üßä Enneper Surface")
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

X = (np.cos(u) * (np.cos(u / 2) * (np.sqrt(2) - np.sin(v)) + np.sin(u / 2) * np.cos(v)))
Y = (np.sin(u) * (np.cos(u / 2) * (np.sqrt(2) - np.sin(v)) + np.sin(u / 2) * np.cos(v)))
Z = -np.sin(u / 2) * (np.sqrt(2) - np.sin(v)) + np.cos(u / 2) * np.cos(v)

plot_surface(X, Y, Z, "üß¨ Klein Bottle Projection")
```

```python
u = np.linspace(0, 12*np.pi, 200)
v = np.linspace(0, 2*np.pi, 50)
u, v = np.meshgrid(u, v)

X = np.sin(u) * (np.exp(np.cos(u)) - 2 * np.cos(4*u) - np.sin(u/12)**5)
Y = np.cos(u) * (np.exp(np.cos(u)) - 2 * np.cos(4*u) - np.sin(u/12)**5)
Z = np.sin(v)

plot_surface(X, Y, Z, "ü¶ã Butterfly Curve Surface", cmap=cm.plasma)
```

```python
u = np.linspace(0, np.pi, 100)
v = np.linspace(0, 2*np.pi, 100)
u, v = np.meshgrid(u, v)

X = (np.cos(u) * np.sin(v)) / (1 + np.sin(u)**2)
Y = (np.sin(u) * np.sin(v)) / (1 + np.sin(u)**2)
Z = np.cos(v)

plot_surface(X, Y, Z, "üòà Boy's Surface Approximation")
```

```python
u = np.linspace(0, np.pi, 100)
v = np.linspace(0, 2*np.pi, 100)
u, v = np.meshgrid(u, v)

X = np.sin(u) * np.cos(v)**2
Y = np.sin(u) * np.sin(v)**2
Z = np.cos(u) * np.sin(2*v)

plot_surface(X, Y, Z, "üí† Roman Surface")
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-1.5, 1.5, 100)
u, v = np.meshgrid(u, v)

a = 1
X = a * np.cosh(v) * np.cos(u)
Y = a * np.cosh(v) * np.sin(u)
Z = v

plot_surface(X, Y, Z, "‚ûï Catenoid (Minimal Surface)")
```
